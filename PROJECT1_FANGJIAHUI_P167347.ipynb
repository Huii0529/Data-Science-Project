### **PART 3 - FUNCTIONS AND CLASSES**



```python
# Parent class
class PetFood:
    def __init__(self, name, category, price, monthly_sales):
        self.name = name
        self.category = category
        self.price = price
        self.monthly_sales = monthly_sales

    def total_sales_revenue(self):
        return self.price * self.monthly_sales - (self.price * 0.35)

    def display_info(self):
        return (
            "Product Formula: {}\n".format(self.name) +
            "Category: {}\n".format(self.category) +
            "Price: MYR{:.2f}\n".format(self.price) +
            "Monthly Sales: {}\n".format(self.monthly_sales) +
            "Total Sales Revenues: MYR{:.2f}".format(self.total_sales_revenue())
        )

# Child class for cat food
class CatFood(PetFood):
    def __init__(self, name, price, type_food, monthly_sales):
        super().__init__(name, "Cat Food", price, monthly_sales)
        self.type_food = type_food

    def display_info(self):
        return (
            super().display_info() +
            "\nType: {}".format(self.type_food)
        )

# Child class for dog food
class DogFood(PetFood):
    def __init__(self, name, price, type_food, monthly_sales):
        super().__init__(name, "Dog Food", price, monthly_sales)
        self.type_food = type_food

    def display_info(self):
        return (
            super().display_info() +
            "\nType: {}".format(self.type_food)
        )
```

```python
# Create instances of CatFood
cat_foods = [
    CatFood("Chicken & Pomegranate", 130, "Dry", 150),
    CatFood("Tuna & Shrimp", 8.5, "Treats", 120),
    CatFood("Herring & Orange", 125, "Dry", 130),
    CatFood("Duck & Cranberry", 8.5, "Wet", 110),
    CatFood("Quail & Coconut", 150, "Dry", 140)
]

# Create instances of DogFood
dog_foods = [
    DogFood("Lamb & Blueberry", 130, "Dry", 160),
    DogFood("Chicken & Pomegranate", 130, "Wet", 150),
    DogFood("Cod & Pumpkin", 10, "Wet", 140),
    DogFood("Salmon & Spelts", 100, "Treats", 130),
    DogFood("Duck & Coconut", 12, "Wet", 120)
]

# Display info for cat food
print("Top 4 Monthly Sales Revenue's Cat Food:\n")
for cat in cat_foods:
    print(cat.display_info())
    print("\n")

# Display info for dog food
print("-----------------------------------------")
print("Top 4 Monthly Sales Revenue's Dog Food:\n")
for dog in dog_foods:
    print(dog.display_info())
    print("\n")
```

### **PART 5 - NUMPY**

```python
import numpy as np

# First set of arrays
A = np.random.rand(5)
B = np.random.rand(3, 3)
C = np.random.rand(2, 4)
D = np.random.rand(4, 2)
E = np.random.rand(5, 5)

# Second set of arrays
F = np.random.rand(5)
G = np.random.rand(3, 3)
H = np.random.rand(2, 4)
I = np.random.rand(4, 2)
J = np.random.rand(5, 5)

# Print each array to display its contents
print("Array A (1D, size 5):\n{}".format(A))
print("\nArray B (2D, 3x3):\n{}".format(B))
print("\nArray C (2D, 2x4):\n{}".format(C))
print("\nArray D (2D, 4x2):\n{}".format(D))
print("\nArray E (2D, 5x5):\n{}".format(E))
print("\nArray F (1D, size 5):\n{}".format(F))
print("\nArray G (2D, 3x3):\n{}".format(G))
print("\nArray H (2D, 2x4):\n{}".format(H))
print("\nArray I (2D, 4x2):\n{}".format(I))
print("\nArray J (2D, 5x5):\n{}".format(J))
```

```python
print("\n--- 1. Perform operation on single array ---")

# Sum of all elements
print("\nSum of elements in A: {}".format(np.sum(A)))

# Mean of all elements
print("Mean of elements in B: {}".format(np.mean(B)))

# Maximum element
print("Maximum element in C: {}".format(np.max(C)))

# Minimum element
print("Minimum element in D: {}".format(np.min(D)))

# Standard deviation
print("Standard deviation of E: {}".format(np.std(E)))

# Reshape array A to (1, 5)
reshaped_A = A.reshape(1, 5)
print("\nReshaped A (1x5):\n{}".format(reshaped_A))

# Transpose array C
transposed_C = C.T
print("\nTransposed C (4x2):\n{}".format(transposed_C))

# Sort array A
sorted_A = np.sort(A)
print("\nSorted A: {}".format(sorted_A))

# Find unique elements
rand_arr_with_dups = np.array([1, 2, 2, 3, 4, 4, 5])
unique_elements = np.unique(rand_arr_with_dups)
print("\nUnique elements in [1, 2, 2, 3, 4, 4, 5]: {}".format(unique_elements))

# Square root of elements in A
sqrt_A = np.sqrt(A)
print("\nSquare root of elements in A: {}".format(sqrt_A))

# Absolute value of elements
abs_val_A = np.abs(A - 0.5)
print("\nAbsolute values of (A - 0.5): {}".format(abs_val_A))

# Round elements of A to 2 decimal places
rounded_A = np.round(A, 2)
print("\nRounded A (2 decimal places): {}".format(rounded_A))
```

```python
print("\n--- 2. Perform operation on two arrays of same size ---")

# Addition of A and F
print("\nElement-wise addition (A + F):\n{}".format(np.add(A, F)))

# Subtraction of G from B
print("\nElement-wise subtraction (B - G):\n{}".format(np.subtract(B, G)))

# Multiplication of C and H
print("\nElement-wise multiplication (C * H):\n{}".format(np.multiply(C, H)))

# Division of E by J
epsilon = 1e-8
print("\nElement-wise division (E / J):\n{}".format(np.divide(E, J + epsilon)))

# Power (F to the power of A)
print("\nElement-wise power (F ** A):\n{}".format(np.power(F, A)))

# Comparison (B > G)
print("\nElement-wise comparison (B > G):\n{}".format(np.greater(B, G)))

# Minimum of C and H
print("\nElement-wise minimum (np.minimum(C, H)):\n{}".format(np.minimum(C, H)))

print("\n--- 3. Perform operation on two arrays of different size (Dot Products) ---")
# Dot product of C (2x4) and D (4x2)
dot_product_CD = np.dot(C, D)
print("\nDot product of C and D (2x4 @ 4x2 = 2x2):\n{}".format(dot_product_CD))
```
